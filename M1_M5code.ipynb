```python
import yfinance as yf
import pandas as pd
import numpy as np
from itertools import product
```

```python
ticker = "BTC-USD"
df = yf.download(ticker, start="2018-12-01", end="2025-12-31")
df = df[['Close']].dropna()
```

```python
df['log_ret'] = np.log(
    df['Close'] / df['Close'].shift(1)
)
df = df.dropna()

q30 = df['log_ret'].quantile(0.30)
q70 = df['log_ret'].quantile(0.70)

def categorize(x):

    if x < q30:
        return -1

    elif x > q70:
        return 1

    else:
        return 0

df['state'] = df['log_ret'].apply(categorize)

df['s1'] = df['state'].shift(4)
df['s2'] = df['state'].shift(3)
df['s3'] = df['state'].shift(2)
df['s4'] = df['state'].shift(1)
df['s5'] = df['state']

df = df.dropna()

state_cols = ['s1','s2','s3','s4','s5']

df[state_cols] = df[state_cols].astype(int)

df['quad'] = list(zip(
    df['s1'],
    df['s2'],
    df['s3'],
    df['s4']
))

order = []

for s4 in [-1,0,1]:
    for s3 in [-1,0,1]:
        for s2 in [-1,0,1]:
            for s1 in [-1,0,1]:

                order.append(
                    (s1,s2,s3,s4)
                )

state_counts = (
    df['quad']
    .value_counts()
)

state_probs = (
    state_counts /
    state_counts.sum()
)
print(state_probs)


state_probs = (
    state_probs
    .reindex(order)
    .fillna(0)
)

transition_counts = df.groupby(
    ['quad', 's5']
).size().unstack(fill_value=0)

transition_probs = transition_counts.div(
    transition_counts.sum(axis=1),
    axis=0
)
transition_probs = (
    transition_probs
    .reindex(order)
    .fillna(0)
)

transition_probs = transition_probs[
    [-1,0,1]
]
transition_probs

ret_table = df.groupby(
    ['quad', 's5']
)['log_ret'].mean().unstack(fill_value=0)

ret_table = (
    ret_table
    .reindex(order)
    .fillna(0)
)

# Ensure column order
ret_table = ret_table[
    [-1,0,1]
]
ret_table

final_ret = (
    transition_probs *
    ret_table
).sum(axis=1)


final_table = pd.concat(
    [
        state_probs.rename(
            "State_Prob"
        ),

        transition_probs.add_prefix(
            "Prob_"
        ),

        ret_table.add_prefix(
            "Ret_"
        ),

        final_ret.rename(
            "Final_Ret"
        )
    ],
    axis=1
)
final_table = (
    final_table * 100
)
prob_cols = [

    c for c
    in final_table.columns

    if "Prob" in c
]

ret_cols = [

    c for c
    in final_table.columns

    if "Ret" in c
]

final_table[prob_cols] = (
    final_table[prob_cols]
    .round(2)
)

final_table[ret_cols] = (
    final_table[ret_cols]
    .round(3)
)
'''print("\n=== FINAL TABLE ===")
print(final_table)'''
final_table

final_table.to_csv('M4_BTC-USD.csv')
print("M4_BTC-USD.csv has been saved to your Colab environment.")
```

```python
# =========================================================
# INSTALL & IMPORTS
# =========================================================
#!pip install yfinance pandas numpy

import yfinance as yf
import pandas as pd
import numpy as np

# =========================================================
# 1. DOWNLOAD DATA
# =========================================================
ticker = "BTC-USD"
df = yf.download(ticker, start="2018-12-01", end="2025-12-31")
df = df[['Close']].dropna()

# =========================================================
# 2. LOG RETURNS
# =========================================================
df['log_ret'] = np.log(
    df['Close'] / df['Close'].shift(1)
)

df = df.dropna()

# =========================================================
# 3. DISCRETIZATION
#
# <30%  -> -1
# 30-70 -> 0
# >70%  -> 1
# =========================================================
q30 = df['log_ret'].quantile(0.30)
q70 = df['log_ret'].quantile(0.70)

def categorize(x):

    if x < q30:
        return -1

    elif x > q70:
        return 1

    else:
        return 0

df['state'] = df['log_ret'].apply(categorize)

# =========================================================
# 4. CREATE STATES
#
# s1 = 1 day ago
# s2 = current day
# =========================================================
df['s1'] = df['state'].shift(1)
df['s2'] = df['state']

df = df.dropna()

df[['s1','s2']] = df[
    ['s1','s2']
].astype(int)

# =========================================================
# 5. CREATE STATE PAIR
#
# (s1) -> s2
# =========================================================
df['pair'] = list(zip(
    df['s1']
))

# =========================================================
# 6. CUSTOM ORDER
#
# s1 changes fastest
# =========================================================
order = [
    (-1,),
    (0,),
    (1,)
]

# =========================================================
# 7. STATE PROBABILITIES
#
# P(s1)
# =========================================================
state_counts = (
    df['pair']
    .value_counts()
)

state_probs = (
    state_counts /
    state_counts.sum()
)

state_probs = (
    state_probs
    .reindex(order)
    .fillna(0)
)

# =========================================================
# 8. TRANSITION COUNTS
#
# (s1) -> s2
# =========================================================
transition_counts = df.groupby(
    ['pair', 's2']
).size().unstack(fill_value=0)

# =========================================================
# 9. TRANSITION PROBABILITIES
#
# P(s2 | s1)
# =========================================================
transition_probs = transition_counts.div(
    transition_counts.sum(axis=1),
    axis=0
)

transition_probs = (
    transition_probs
    .reindex(order)
    .fillna(0)
)

# Ensure column order
transition_probs = transition_probs[
    [-1,0,1]
]

# =========================================================
# 10. EXPECTED RETURNS
#
# E[ret | s1,s2]
# =========================================================
ret_table = df.groupby(
    ['pair', 's2']
)['log_ret'].mean().unstack(fill_value=0)

ret_table = (
    ret_table
    .reindex(order)
    .fillna(0)
)

# Ensure column order
ret_table = ret_table[
    [-1,0,1]
]

# =========================================================
# 11. FINAL EXPECTED RETURN
#
# E[R | s1]
# =========================================================
final_ret = (
    transition_probs *
    ret_table
).sum(axis=1)

# =========================================================
# 12. CREATE FINAL TABLE
# =========================================================
final_table = pd.concat(
    [
        state_probs.rename(
            "State_Prob"
        ),

        transition_probs.add_prefix(
            "Prob_"
        ),

        ret_table.add_prefix(
            "Ret_"
        ),

        final_ret.rename(
            "Final_Ret"
        )
    ],
    axis=1
)

# =========================================================
# 13. CLEAN INDEX FORMAT
# =========================================================
final_table.index = [
    f"{a}"
    for (a,)
    in final_table.index
]

# =========================================================
# 14. CONVERT TO %
# =========================================================
final_table = (
    final_table * 100
)

# =========================================================
# 15. ROUNDING
#
# Probabilities -> 2 decimals
# Returns       -> 3 decimals
# =========================================================
prob_cols = [

    c for c
    in final_table.columns

    if "Prob" in c
]

ret_cols = [

    c for c
    in final_table.columns

    if "Ret" in c
]

final_table[prob_cols] = (
    final_table[prob_cols]
    .round(2)
)

final_table[ret_cols] = (
    final_table[ret_cols]
    .round(3)
)

# =========================================================
# DISPLAY
# =========================================================
print("\n=== FINAL TABLE ===")
print(final_table)
final_table.to_csv('M1_BTC-USD.csv')
print("M1_BTC-USD.csv has been saved to your Colab environment.")
# =========================================================
# VALIDATION CHECKS
# =========================================================

print("\n=== CHECK 1 ===")
print(
    "State probabilities sum:"
)

print(
    final_table[
        'State_Prob'
    ].sum()
)

print("\n=== CHECK 2 ===")
print(
    "Transition row sums:"
)

print(
    transition_probs
    .sum(axis=1)
)

print("\n=== CHECK 3 ===")
print(
    "Total states:"
)

print(
    len(final_table)
)
```
